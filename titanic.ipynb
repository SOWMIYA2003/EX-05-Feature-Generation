```python
import pandas as pd
df=pd.read_csv("titanic_dataset.csv")
df
```

```python
df.drop("Name",axis=1,inplace=True)
```

```python
df
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
df.isnull().sum()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
oe=OrdinalEncoder()
oe.fit_transform(df[["Embarked"]])
```

```python
embark=['S','C','Q']
enc=OrdinalEncoder(categories=[embark])
```

```python
enc
```

```python
enc.fit_transform(df[['Embarked']])
```

```python
df1=df.copy()
df1["Embarked"]=enc.fit_transform(df[["Embarked"]])
df1
```

```python
pip install category_encoders
```

```python
from category_encoders import BinaryEncoder
be=BinaryEncoder()
newdata=be.fit_transform(df1["Sex"])
newdata
```

```python
df2=df1.copy()
df2["Sex"]=be.fit_transform(df1[["Sex"]])
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df3=pd.DataFrame(scaler.fit_transform(df2),columns=['Passenger','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])
```

```python
df3
```

```python
import matplotlib.pyplot as plt
ax=df2.plot.scatter(x="Age",y="Survived",marker="*",label="Before-Scaling",color="red")
df3.plot.scatter(x="Age",y="Survived",marker="+",label="After-Scaling",ax=ax)
plt.axhline(0, color='purple',alpha=0.5)
plt.axvline(0, color='purple',alpha=0.5);
plt.show()
```

```python
import matplotlib.pyplot as plt
ax=df2.plot.scatter(x="Age",y="Fare",marker="*",label="Before-Scaling",color="red")
df3.plot.scatter(x="Age",y="Fare",marker="+",label="After-Scaling",ax=ax)
plt.axhline(0, color='purple',alpha=0.5)
plt.axvline(0, color='purple',alpha=0.5);
plt.show()
```

```python
import matplotlib.pyplot as plt
ax=df2.plot.scatter(x="Age",y="Embarked",marker="*",label="Before-Scaling",color="red")
df3.plot.scatter(x="Age",y="Embarked",marker="+",label="After-Scaling",ax=ax)
plt.axhline(0, color='purple',alpha=0.5)
plt.axvline(0, color='purple',alpha=0.5);
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
Stdscaler=StandardScaler()
df4=pd.DataFrame(Stdscaler.fit_transform(df2),columns=['Passenger','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])
df4
```

```python
import matplotlib.pyplot as plt
axstd=df2.plot.scatter(x="Age",y="Fare",marker="*",label="Before-Scaling",color="red")
df4.plot.scatter(x="Age",y="Fare",marker="+",label="After-Scaling",ax=axstd)
plt.axhline(0, color='orange',alpha=0.5)
plt.axvline(0, color='orange',alpha=0.5);
plt.show()
```

```python
from sklearn.preprocessing import MaxAbsScaler
maxabsscaler=MaxAbsScaler()
df5=pd.DataFrame(maxabsscaler.fit_transform(df2),columns=['Passenger','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])
df5
```

```python
axmax=df2.plot.scatter(x="Age",y="Fare",marker="*",label="Before-Scaling",color="red")
df5.plot.scatter(x="Age",y="Fare",marker="+",label="After-Scaling",ax=axmax)
plt.axhline(0, color='blue',alpha=0.5)
plt.axvline(0, color='blue',alpha=0.5);
plt.show()
```

```python
axmax=df2.plot.scatter(x="Embarked",y="Fare",marker="*",label="Before-Scaling",color="red")
df5.plot.scatter(x="Embarked",y="Fare",marker="+",label="After-Scaling",ax=axmax)
plt.axhline(0, color='blue',alpha=0.5)
plt.axvline(0, color='blue',alpha=0.5);
plt.show()
```
